06 - Faça um Programa que peça o raio de um círculo, calcule e mostre sua área.

```python
import math
raio = float(input("Informe o raio do círculo: "))

area_circulo = round(math.pi * pow(raio,2),2)

print("A área do Circulo é: ", area_circulo)
```

9 - Faça um Programa que peça a temperatura em graus Fahrenheit, transforme e mostre a temperatura em graus Celsius.

14 - João Papo-de-Pescador, homem de bem, comprou um microcomputador para controlar o rendimento diário de seu trabalho. João precisa que você faça um programa que leia a variável peso (peso de peixes) e calcule o excesso. Gravar na variável excesso a quantidade de quilos além do limite e na variável multa o valor da multa que João deverá pagar. Imprima os dados do programa com as mensagens adequadas. 

```python
peso_pescado = int(input('Informe quantos kg foram pescados: '))

peso_limite = 50

multa = (peso_pescado - peso_limite) * 4

print(f'Peso Pescado: {peso_pescado} multa:R${0 if peso_pescado <=50 else multa}')
```

15 - Faça um Programa que pergunte quanto você ganha por hora e o número de horas trabalhadas no mês. Calcule e mostre o total do seu salário no referido mês, sabendo-se que são descontados 11% para o Imposto de Renda, 8% para o INSS e 5% para o sindicato, faça um programa que nos dê:

    salário bruto.
    quanto pagou ao INSS.
    quanto pagou ao sindicato.
    o salário líquido.
    calcule os descontos e o salário líquido, conforme a tabela abaixo:

    + Salário Bruto : R$
    - IR (11%) : R$
    - INSS (8%) : R$
    - Sindicato ( 5%) : R$
    = Salário Liquido : R$

```python
val_hora = float(input('Informa o salário Hora: '))
qtd_horas = float(input('Informe o número de horas trabalhadas: '))

salario_bruto = val_hora * qtd_horas
imposto_renda = salario_bruto*0.11
previdencia = salario_bruto*0.14
sindicato = val_hora*8
salario_liquido = salario_bruto - (imposto_renda + previdencia + sindicato)

print(f'Salario Bruto: R${salario_bruto}\nIR: R${imposto_renda}\nINSS: R${previdencia}\nSindicato: R${sindicato}\nSalário Líquido:R${salario_liquido}')
```

16 - Faça um programa para uma loja de tintas. O programa deverá pedir o tamanho em metros quadrados da área a ser pintada. Considere que a cobertura da tinta é de 1 litro para cada 3 metros quadrados e que a tinta é vendida em latas de 18 litros, que custam R$ 80,00. Informe ao usuário a quantidades de latas de tinta a serem compradas e o preço total.

```python
import math

area_a_cobrir = float(input('Qual o tamanho do quarto: '))

qtd_nes = area_a_cobrir / 3

qtd_latas = qtd_nes / 18

if area_a_cobrir % 54 == 0:
	qtd_latas = area_a_cobrir / 54
else: 
	qtd_latas = int(area_a_cobrir / 54) + 1

print(f'Area do quarto: {area_a_cobrir} Latas à usar: {qtd_latas} Valor da compra: {qtd_latas*80}')
```
